```python
def retrieve_text(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    html_page = requests.get(url, headers=headers)
    soup = BeautifulSoup(html_page.content, 'html.parser')
    para = soup.find("div", {"class": "gsc_oci_value", "id" : "gsc_oci_descr"})
    return para.get_text()
```

```python
df["description"] = df["URL"].apply(retrieve_text)
```

```python
df
```

```python
for idx, row in df.iterrows():
    file_text = \
f"""
+++
title = "{row['Name']}"
author = "{row['Authors']}"
description = "{row['description']}"
date = "{row['Year']}-02-28"
journal = "{row['Journal']}"
tags = [
    "publication",
]
+++

### Abstract:

{row['description']}

### URL:

{row['URL']}
"""

    if row["Journal"]:
        file_text += f"\n### Published @ {row['Journal']}\n"
    with open(f"content/publications/pub{idx:02d}.md", "w") as outfile:
        outfile.write(file_text)
```
